```python
from Crypto.Cipher import AES
from Crypto.Util.number import *
from Crypto.Util.Padding import pad, unpad


import hashlib
import json
from os import urandom

FLAG = b'crypto{?????????????????????????????????????}'

P = 13322168333598193507807385110954579994440518298037390249219367653433362879385570348589112466639563190026187881314341273227495066439490025867330585397455471
N = 30

def load_matrix(fname):
    data = open(fname, 'r').read().strip()
    rows = [list(map(int, row.split(' '))) for row in data.splitlines()]
    return Matrix(GF(P), rows)

def decrypt_cipher(secret, ciphertext, iv):
    
    ciphertext = bytes.fromhex(ciphertext)
    iv = bytes.fromhex(iv)
    
    KEY_LENGTH = 128
    k = hashlib.sha256(str(secret).encode()).digest()[:KEY_LENGTH]
    
    cipher = AES.new(k, AES.MODE_CBC, iv)
    plaintext = cipher.decrypt(ciphertext)

    return plaintext
    
    
def get_jordan_blocks(J):
    blocks = []
    i = 0
    while i < J.nrows():
        j = i
        while j < J.nrows() and J[j, j] == J[i, i]:
            j += 1
            
        blocks.append(J[range(i, j), range(i, j)])
        i = j
        
    return blocks
    
    
G = load_matrix("files\\matrix_reloaded\\generator.txt")

with open("files\\matrix_reloaded\\flag.enc") as f:
    flag_params = json.load(f)

    
with open("files\\matrix_reloaded\\output.txt") as f:
    output_params = json.load(f)

v = vector(GF(P), output_params["v"])
w = vector(GF(P), output_params["w"])
    
# P_transformation * J * P_transformation^-1 = G
J, P_transformation = G.jordan_form(transformation=True)
 

# J^k * P^-1 * v = P^-1 * w
# X = P^-1 * v
# Y = P^-1 * w
# J^k * X = Y
X = P_transformation.inverse() * v
Y = P_transformation.inverse() * w

jordan_blocks = get_jordan_blocks(J)
# The last eigenvalue size is 2
# jordan_blocks[-1]

e = jordan_blocks[-1][0,0]

def pinv(n):
    return pow(n, P-2, P)

x0, x1 = X[-2], X[-1]
y0, y1 = Y[-2], Y[-1]

k = (y0 - (x0 * (y1 * pinv(x1)))) * pinv(x1) * e * pinv((y1 * pinv(x1)))
k = int(k)
k = k % P

# Verify
H = G^k
print(w == H*v)

print(decrypt_cipher(k, flag_params["ciphertext"], flag_params["iv"]))
```
